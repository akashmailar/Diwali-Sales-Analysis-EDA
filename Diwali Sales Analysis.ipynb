# Diwali Sales  Analysis - EDA :-

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

### To avoid encoding error, use encoding='Latin1'.

```python
df = pd.read_csv("DiwaliSalesData.csv", encoding='Latin1')
df.head()
```

### To check the Null input data.

```python
df.info()
```

### Deleting the NaN Columns, cleaning the data.

```python
df.drop(['Status', 'unnamed1'], axis=1, inplace=True)
```

```python
df.info()
```

### To check the total Null values present in the data.

```python
df.isna().sum()
```

### To Drop the Null values columns.

```python
df.dropna(inplace=True)
```

```python
df.isna().sum()
```

```python
df.head()
```

```python
df['Product_Category'].unique()
```

```python
df.shape
```

### Statistical Data :-

```python
df.describe()
```

```python
df[['Age', 'Orders', 'Amount']].describe()
```

```python
df['Amount'] = df['Amount'].astype('int')
```

## Exploratory Data Analysis :-

## Count of Customers Based on `Gender` :-

```python
df["Gender"].value_counts()
```

```python
color = sns.color_palette("Set2")
p = sns.countplot(data=df, x='Gender', palette=color)

for i in p.containers:
    p.bar_label(i)

plt.title("Count of Customers Based on Gender")
plt.show()
```

## Total `Amount` Spent Based on `Gender` :- 

```python
sales_grouped = df.groupby('Gender')['Amount'].sum().reset_index()
sales_grouped
```

```python
p1 = sns.barplot(data=sales_grouped, x='Gender', y='Amount', palette=color)

plt.title("Maximum Amount Spent Based on Gender")
plt.show()
```

### From the above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men.

```python
df.columns
```

## Count of Customers Based on `Age Group` :-

```python
p1 = sns.countplot(data=df, x='Age Group', palette=color)

for i in p1.containers:
    p1.bar_label(i)

plt.title("Count of Customers Based on Age Group")
plt.show()
```

## Count of Customers Based on `Age Group` and `Gender` :-

```python
p1 = sns.countplot(data=df, x='Age Group', hue='Gender')

for i in p1.containers:
    p1.bar_label(i)

plt.title("Count of Customers Based on Age Group and Gender")
plt.show()
```

## Total `Amount` Vs. `Age Group` and `Gender` :-

```python
total_agegroup = df.groupby(['Age Group', 'Gender'])['Amount'].sum().reset_index()
total_agegroup
```

```python
sns.barplot(data=total_agegroup, x='Age Group', y='Amount', hue='Gender')

plt.title("Total Amount Vs. Age Group and Gender")
plt.show()
```

### From the above graphs we can see that most of the buyers are of `Age Group` between 26-36 years and mostly females. 

```python
df.columns
```

## Top 10 `States` with Most `Orders` :-

```python
Orders_State_grouped = df.groupby('State')['Orders'].sum().reset_index()
Orders_State_grouped.sort_values(by='Orders', ascending=False, inplace=True, ignore_index=True)
```

```python
top_10 = Orders_State_grouped.head(10)
top_10
```

```python
p2 = sns.barplot(data=top_10, x='State', y='Orders', palette=color)

for i in p2.containers:
    p2.bar_label(i)

plt.xticks(rotation=90)
plt.title("Top 10 States with Most Orders")
plt.show()
```

## Top 5 `States` with Most `Amount` Spent :-

```python
State_Amount_grouped = df.groupby('State')['Amount'].sum().reset_index()
State_Amount_grouped.sort_values(by='Amount', ascending=False, inplace=True, ignore_index=True)
top_5 = State_Amount_grouped.head(5)
top_5
```

```python
p3 = sns.barplot(data=top_5, x='State', y='Amount', palette=color)

plt.title("Top 5 States with Most Amount Spent")

plt.show()
```

### From the above graphs we can see that most of the `Orders` are from Uttar Pradesh, Maharashtra and Karnataka respectively. And also, Maximum `Amount` Spent is from UP, Maharashtra and Karnataka.

## Count of Customers Based On `Marital Status` :-

```python
df['Marital_Status'].unique()
```

### Changing 0 and 1 to 'Married' and 'Un-Married'

```python
def marital_status(value):
    if value == 0:
        return 'Married'
    else:
        return 'Un-Married'

df['Marital_Status'] = df['Marital_Status'].apply(marital_status)
```

```python
df['Marital_Status'].unique()
```

```python
df['Marital_Status'].value_counts()
```

```python
p4 = sns.countplot(data=df, x='Marital_Status', palette=color)

for i in p4.containers:
    p4.bar_label(i)

p4.set(xlabel="Marital Status")
plt.title("Count of Customers Based On Marital Status")
plt.show()
```

## Count of Customers Based On `Marital Status` and `Gender` :-

```python
p5 = sns.countplot(data=df, x='Marital_Status', hue='Gender')

for i in p5.containers:
    p5.bar_label(i)

p5.set(xlabel='Marital Status')
plt.title("Count of Customers Based On Marital Status and Gender")
plt.show()
```

## Maximum Amount Spent Based on `Marital Status` and `Gender` :-

```python
Amount_marital_grouped = df.groupby(['Marital_Status', 'Gender'])['Amount'].sum().reset_index()
Amount_marital_grouped
```

```python
p6 = sns.barplot(data=Amount_marital_grouped, x='Marital_Status', y='Amount', hue='Gender')

p6.set(xlabel='Marital Status')
plt.title("Maximum Amount Spent Based on Marital Status and Gender")
plt.show()
```

### From the above graphs we can see that most of the buyers are married (women) and also they have high purchasing power.

## Total `Orders` Based On `Product Category` :-

```python
Orders_product_grouped = df.groupby('Product_Category')['Orders'].sum().reset_index()
Orders_product_grouped.sort_values(by='Orders', ascending=False, inplace=True, ignore_index=True)
Orders_product_grouped
```

## Maximum `Amount` Spent Based On `Product Category` :-

```python
Amount_product_grouped = df.groupby('Product_Category')['Amount'].sum().reset_index()
Amount_product_grouped.sort_values(by='Amount', ascending=False, inplace=True, ignore_index=True)
Amount_product_grouped
```

```python
plt.figure(figsize=(20,5))
p10 = sns.barplot(data=Amount_product_grouped, x='Product_Category', y='Amount', palette=color)

plt.xticks(rotation = 45)
plt.show()
```

### From the above graphs we can see that most of the sold products are from Food, Clothing & Apparel and Electronics & Gadgets.
